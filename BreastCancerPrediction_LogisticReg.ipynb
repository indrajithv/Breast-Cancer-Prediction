```python
import numpy as np
import sklearn.datasets
breastcancer = sklearn.datasets.load_breast_cancer()
x = breastcancer.data
y = breastcancer.target
print(x)
print(y)
import pandas as pd
data = pd.DataFrame(breastcancer.data,columns = breastcancer.feature_names)
data.head()
```

```python
data["class"] = breastcancer.target
print(data['class'].value_counts())
print(breastcancer.target_names)
data.groupby("class").mean()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1)
print(y.shape,y_train.shape,y_test.shape)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,stratify = y,random_state = 1)
print(y.mean(),y_train.mean(),y_test.mean())
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
predict_train = classifier.predict(x_train)
accuracy_train = accuracy_score(y_train,predict_train)
print('Accuracy of training data is',accuracy_train)
predict_test = classifier.predict(x_test)
accuracy_test = accuracy_score(y_test,predict_test)
print('Accuracy of testing data is',accuracy_test)
```
